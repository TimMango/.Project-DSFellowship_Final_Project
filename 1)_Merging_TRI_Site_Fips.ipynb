```python
#Import Packages

import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 5000)
pd.set_option('display.width', 1000)
```

```python
#Import Data and observe columns

df = pd.read_csv('TRI_2017_US_copy.csv')
df.head(1)
```

```python
#These are the Fips codes for each column row

df1 = pd.read_csv('FIPS_Round1_Final.csv')
df2 = pd.read_csv('FIPS_Round2_Final.csv')
df3 = pd.read_csv('FIPS_Round3_Final.csv')

fips_code_df = pd.concat([df1, df2, df3], ignore_index=True)
fips_code_df.drop(columns=['API_Time', 'Unnamed: 0'], inplace=True)
```

```python
##This creates Unique column for mergin fips_code_df
fips_code_df['Merging_Column'] = fips_code_df['Latitude'] + fips_code_df['Longitude']
```

```python
#These are variables of interest from the TRI report dataframe
vars_of_interest_df = df[['1. YEAR', '3. FRS ID', '5. STREET ADDRESS', '12. LATITUDE', '13. LONGITUDE', '15. INDUSTRY SECTOR CODE', '16. INDUSTRY SECTOR', '30. CHEMICAL', '31. CAS #/COMPOUND ID', '36. METAL CATEGORY', '54. ON-SITE RELEASE TOTAL', '112. PARENT CO NAME', '113. PARENT CO DB NUM']]
```

```python
#Create unique merging column from TRI report vars of interest

vars_of_interest_df['Merging_Column'] = (vars_of_interest_df['12. LATITUDE'] + vars_of_interest_df['13. LONGITUDE'])
```

```python
#Aggregate vars of interest by merging column and chemical
#This gives chemical count for unique latitude and longitudes

unique_merging_df = vars_of_interest_df.groupby(['Merging_Column','30. CHEMICAL']).aggregate(sum)
```

```python
#Aggregate fips codes to unique IDs

unique_merging_fips = fips_code_df.groupby('Merging_Column').aggregate(np.median)
```

```python
#Combine unique locations with unique FIPs codes

unique_merged_df = pd.merge(unique_merging_df, unique_merging_fips, left_index=True, right_index=True, how='left')
```

```python
#This dataframe shows chemicals at each unique location

unique_merged_df.head()
```

```python
final_df = unique_merged_df
final_df.reset_index(inplace=True)
```

```python
#Get Target Variable Ready for export

export_2017_target = final_df[['30. CHEMICAL', 'FIPS']]
export_2017_target['Target'] = 1
```
